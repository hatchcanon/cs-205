```python
import nltk
# Read in the text as a string
textRaw = open("res/hg.txt").read()
textRaw
```

```python
# Use Python Regular Expression library
import re

match = r'\d+ \| P a g e'
textRaw = re.sub(match, '', textRaw)

match = r'The Hunger Games - Suzanne Collins'
textRaw = re.sub(match, '', textRaw)
textRaw
```

```python
def writeWordFrequency(words, title):
    import json
    from collections import Counter

#Count the frequency of each word in the array
    freq = Counter()
    for w in words:
        freq[w] += 1
    freq_table = dict(freq.most_common(30))
    print (freq.most_common(30))
# Create an object containing all of our data
    outData = { "title": title, "freq": freq_table }

# Write the object out to a file
    with open("res/" + title + ".json", "w") as f:
        json.dump(outData, f, indent = 2)
```

```python
# Import the NLTK tokenize for word_tokenize
from nltk.tokenize import word_tokenize

# Use it on raw text
words = word_tokenize( textRaw )

# Output our result
writeWordFrequency(words, "tokenize")
```

```python
# Using standard Python to lowercase words:
lowerCaseWords = []
for word in words:
    lowerCaseWords.append(word.lower())
words = lowerCaseWords 

# Output our result
writeWordFrequency(words, "lowercase")
```

```python
# Using standard Python to remove non-words
cleanedWords = []
for word in words:
    if word.isalpha():
        cleanedWords.append(word)
words = cleanedWords

# Output our result
writeWordFrequency(words, "words-only")
```

```python
# Using NLTK corpus library for stopwords
from nltk.corpus import stopwords


# Using standard Python with stopwords
cleanedWords = []
for word in words:
    if word not in stopwords.words("english"):
        cleanedWords.append(word)
words = cleanedWords

# Output our result
writeWordFrequency(words, "stopwords")
```

```python
# Use NLTK stem library for SnowballStemmer
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")
stemmedWords = []
for word in words:
    stemmedWords.append( stemmer.stem(word) )
words = stemmedWords

# Output our result
writeWordFrequency(words, "stem")
```
