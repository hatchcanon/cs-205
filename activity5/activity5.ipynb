```python
df = pd.read_csv("uiuc_demographics_2005_15.csv")
df
```

```python
df = df.dropna()
df ["Degree"] = df ["Degree"].map(str.strip)
```

```python
df = df[ (df["Fall"] == 2005) | (df["Fall"] == 2015)]
df
```

```python
group = df.groupby(["Major Name", "Fall"]).sum()
del group["Term Code"]
del group["Major Code"]
del group["Dept."]
del group["Male"]
del group["Unknown"]
group
```

```python
group ["Female%"] = group ["Female"] / group["Total"]
group
```
