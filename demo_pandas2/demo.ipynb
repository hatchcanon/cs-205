```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("uiuc_demographics_2005_15.csv")
df
```

```python
df["Total"].sum()
```

```python
df = df.dropna()
df["Degree"].sum()
```

```python
df["Degree"] = df["Degree"].map(str.strip)
```

```python
df["Degree"].unique()
```

```python
grp = df.groupby ( ["Major Name", "Fall"])
grp.count()
```

```python
grp.count()
```

```python
grp.agg({
    "Total": np.sum,
    "Male": np.sum,
    "Female": np.sum
})
```

```python
df[ (df["Degree"] == "BSLAS") | (df["Degree"] == "BALAS")]
```

```python
df_ug = df[ df["Degree"].str.startswith("B")]
df_ug = df_ug[
    (df_ug["Fall"] == 2015) |
    (df_ug["Fall"] == 2005)
]
df_ug
```

```python
grp = df_ug.groupby( ["Major Name", "Fall"])
df_final= grp.agg({
    "Total": np.sum,
    "Male": np.sum,
    "Female": np.sum
})
df_final
```
