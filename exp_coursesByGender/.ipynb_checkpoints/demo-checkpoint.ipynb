```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("res/uiuc_demographics_2005_15.csv")
df
```

```python
df = df[ df["Fall"] == 2015]
```

```python
grp = df.groupby("College")
grp.count()
```

```python
df2 = grp.agg({
    "Female": np.sum,
    "Total": np.sum,
})
df2["%Female"] = df2["Female"] / df2["Total"]
df2
```

```python
df2.to_csv("2015uiuc.csv")
```
